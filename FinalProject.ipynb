# data cleaning & pre-processing

```python
df = pd.read_csv('cape.csv')
```

```python
# some data are actually "No CAPEs submitted". Get rid of those dirty data
df = df[df['Course'] != 'No CAPEs submitted']
```

```python
def extract_quarter(term):
    if term[:2] == 'FA':
        return 'Fall'
    elif term[:2] == 'WI':
        return 'Winter'
    elif term[:2] == 'SP':
        return 'Spring'
    elif term[:2] == 'S1' or term[:2] == 'S2' or term[:2] == 'S3' or term[:2] == 'SU':
        return 'Summer'
    else:
        return np.nan

def extract_year(term):
    return int('20' + term[2:])
```

```python
def extract_avg_letter_grade(grade):
    if type(grade) == str:
        return grade.split()[0]
    else:
        return np.nan
```

```python
def extract_avg_gpa(grade):
    if type(grade) == str:
        return float(grade.split()[1][1:-1])
    else:
        return np.nan
```

```python
def extract_percentages(datum):
    if type(datum) == str:
        return float(datum[:4]) / 100
    else:
        return np.nan 
```

```python
def extract_dept_name(dept):
    return dept.split('-')[1].strip()

def extract_dept_code(dept):
    return dept.split('-')[0].strip()
```

```python
def extract_course_code(course):
    return course.split('-')[0].strip()

def extract_course_name(course):
    return course.split('-')[1][:-3].strip() # get rid of (A), (B), (1), etc. at the end
```

```python
# Some columns contain several different information. Split them to different columns here
df['Quarter'] = df['Term'].apply(extract_quarter)
df['Year'] = df['Term'].apply(extract_year)
df['Avg Letter Grade Received'] = df['Avg Grade Received'].apply(extract_avg_letter_grade)
df['Avg GPA Received'] = df['Avg Grade Received'].apply(extract_avg_gpa)
df['Avg Letter Grade Expected'] = df['Avg Grade Expected'].apply(extract_avg_letter_grade)
df['Avg GPA Expected'] = df['Avg Grade Expected'].apply(extract_avg_gpa)
df['Recommend Class'] = df['Rcmnd Class'].apply(extract_percentages)
df['Recommend Instructor'] = df['Rcmnd Instr'].apply(extract_percentages)
df['Department Name'] = df['Department'].apply(extract_dept_name)
df['Department Code'] = df['Department'].apply(extract_dept_code)
df['Course Code'] = df['Course'].apply(extract_course_code)
df['Course Name'] = df['Course'].apply(extract_course_name)

df = df.drop(['Avg Grade Expected', 'Avg Grade Received', 'Term', 'Rcmnd Class', 'Rcmnd Instr', 'Course', 'Department'], axis=1)
```

GPA, study hours, instructor and class recommended rate are data we are focusing on. Therefore we drop all the rows whose entries above is missing.


```python
df.dropna(subset=['Avg GPA Expected','Avg GPA Received', 'Study Hrs/wk', 'Recommend Instructor', 'Recommend Class'], inplace=True)
df.drop(['Instructor','Course Name', 'Avg Letter Grade Received', 'Avg Letter Grade Expected', 'Department Name'], axis=1, inplace=True)
```

In order not to cause privacy problem we drop instructor name and course name.We also drop other duplicated data that is unnecessary.

```python
df['GPA Gap'] =  df['Avg GPA Received'] - df['Avg GPA Expected']
df['CAPE Made Rate'] = df['Evals Made'] / df['Enroll']
```

GPA gap is the difference between GPA received and expected(a negative sign indicates that the GPA expected is lower than that received, and vice versa)
CAPE MADE RATE is the ratio of number of students that submitted CAPE out of that of all students enrolled in one class.

```python
def get_course_category(code):
    num = code.split(' ')[1].strip()
    letter = ['A','B','C','D','E','F','G','J','K','S','L','R','H','M','N','P','T','W','I','X','O','Q']
    while num[len(num) - 1] in letter:
            num = num[:len(num) - 1]
    num = int(num)
    if num < 100:
        return 'lower'
    elif num < 200:
        return 'upper'
    elif num < 300:
        return 'graduate'
    elif num < 400:
        return 'professional'
    else:
        return 'other'
    
```

```python
df['Category'] = df['Course Code'].apply(get_course_category)
```

# Privacy/Ethics Consideration:

We have permission to use those data and UCSD encourages the use of the data of its website for researching purposes therefore using of CAPEs data is allowed.

According to the Safe Harbour Method, we want to avoid the use of name and all personal related information. The datasets contains a lot of professors’ name and specific course number,use of which appears to be a privacy concern so choose to not use professors' name into our analysis since those names are easily decrypted. Regarding to the specific course number, we would only want to keep the department abbreviation at the beginnning of the course number. Since the specific number of the course can be decrypted easily as well. Many courses are only taught by specific professors.

As CAPEs mainly focuses on academic datasets there is little potential biases we found that violates equitable analysis. Limitation: 
During the process of analyzing the correlations among the variables in CAPE, we found some limitations due to the nature of the data from CAPE.
We noticed that in some cases, CAPE puts all different sections with the same course name together under one big class, even though these students attended completely different lectures with different instructors. This could largely affect the results of our analysis in a lot of ways. First of all, the Enroll variable might be very inaccurate for some classes, since the class sizes appear on CAPE could be a few times as large as the actual sizes. As a result, our findings on the correlation between class size and average GPA receive could be biased. Second of all, variables like Avg GPA Received and Recommend Instructor could be about different classes and different instructors, but are all put together, which might also introduce bias.


There could also be other confounding variables that played big roles in influencing our target variables and are not accessible from CAPE. Some of the confounding variables might include lecture times, attendance rate, which could be highly associated with class performance.


All in all, the CAPEs system turns out to be a great reference tool, but still has some bias under certain conditions.
